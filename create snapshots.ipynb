# create snaspshots

```python
%cd {LINUX_REPO_BASE}
!git log --tags --simplify-by-decoration --pretty="format:%ai %d"
```

```python
VERSIONS = {
#     'v2.6.12-rc2':'2005-04-16 15:20:36 -0700',
#     'v2.6.21' : '2007-04-25 20:08:32 -0700',
    'v2.6.31':'2009-09-09 15:13:59 -0700',
    'v3.0':'2011-07-21 19:17:23 -0700',    
#     'v3.11':'2013-09-02 13:46:10 -0700',        
    'v4.0':'2015-04-12 15:12:50 -0700',            
#     'v4.13':'2017-09-03 13:56:17 -0700',
    'v5.0':'2019-03-03 15:21:29 -0800'
}
```

```python
for tag in tqdm(VERSIONS):
    if os.path.exists(f"{SNAPSHOTS_BASE}/{tag}"):
        continue
    !git archive {tag} --format=tar.gz -o {SNAPSHOTS_BASE}/{tag}.tar.gz
    os.makedirs(os.path.join(SNAPSHOTS_BASE, tag), exist_ok=True)
    !tar -xzf "{SNAPSHOTS_BASE}/{tag}.tar.gz" -C "{SNAPSHOTS_BASE}/{tag}"
```
